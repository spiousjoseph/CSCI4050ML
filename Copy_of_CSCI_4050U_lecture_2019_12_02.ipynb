```
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense
```

```
data = [list(range(40, 44)),
        list(range(61, 64)),
        [103],
        list(range(12, 19))]
```

```
data
```

```
ragged_input = tf.ragged.constant(data)
```

```
dense_input = ragged_input[:, :5].to_tensor(default_value=1)
```

```
emb_layer = Embedding(104, 9)
```

```
emb_layer(dense_input).shape
```

```
lstm_layer = LSTM(10, return_sequences=True)
```

```
lstm_layer(emb_layer(dense_input)).shape
```

```
model = Sequential([
    Input(shape=(5,), dtype=tf.int64),
    Embedding(104, 9),
    LSTM(10),
    Dense(2)
])
```

```
model.summary()
```

```
model.compile(loss=tf.losses.sparse_categorical_crossentropy, optimizer='adam')
```
